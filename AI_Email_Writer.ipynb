```python
import time
import gradio as gr
from openai import OpenAI
from google.colab import userdata
```

```python
OPEN_API_KEY = userdata.get('OPENAI_API_KEY')
```

```python
openai = OpenAI(api_key=OPEN_API_KEY)
MODEL="gpt-5-nano"
```

```python
SYSTEM_MESSAGE = f"""
You are an expert business communication specialist and email writer.

Instructions:
1. Create a compelling and relevant subject line if needed.
2. Start with an appropriate greeting.
3. Clearly explain the purpose of the email.
4. Keep the message professional, concise, and engaging.
5. Avoid repetitive wording and unnecessary filler.
6. Do not use excessive bullet points.
7. Keep paragraphs short and readable.
8. End with a professional closing statement.
9. Return only the final email.
10. Maintain a tone throughout the email.

Generate the email now.
"""
```

```python
def email_generator(email_prompt):
    try:
        response = openai.chat.completions.create(
            model=MODEL,
            temperature=0.3,
            messages=[
                {"role": "system", "content": SYSTEM_MESSAGE},
                {"role": "user", "content": email_prompt}
            ]
        )

        return response.choices[0].message.content

    except Exception as e:
        return f"Error: {str(e)}"
```

```python
def form_submit(subject, email_body, tone):
  if not email_body.strip():
    return "Please provide email details."

  email_prompt = f"""
  Write a complete email using the information below:
  Subject: {subject}
  Email Body: {email_body}
  Tone: {tone}
  """
  email = email_generator(email_prompt)
  for i in range(len(email)):
        time.sleep(0.1)
        yield email[:i+1]
```

```python
with gr.Blocks(title="Type a email reason and generate email with tone") as email_form:
  with gr.Row():
    with gr.Column(scale=1):
      subject_input_box = gr.Textbox(lines=1, placeholder="Write a Email Subject")
      emailBody_input_box = gr.Textbox(lines=2, placeholder="Write an email to client about project delay.")
      tone_selection = gr.Dropdown(
          choices=["Professional", "Friendly", "Formal" , "Apology" , "Sales"],
          label="Select a Choice",
          info="Choose one of the options above"
        )
      btn = gr.Button("Generate email" , variant="primary")
    with gr.Column(scale=1):
      output_label = gr.Label("Generated Email")
      output_box = gr.Markdown()
    btn.click(fn=form_submit, inputs=[subject_input_box , emailBody_input_box , tone_selection], outputs=output_box)
```

```python
email_form.launch()
```
